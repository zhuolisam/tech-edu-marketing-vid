```python
#import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#crete example age data within 0 - 50
age = [10,35,12,45,35,10,12,10,26,35,10]
```

```python
#plot distribution of numerical data with histogram
sns.displot(x=age, kde=True)
```

```python
# Set the figure size to (12,6)
plt.figure(figsize=(12,6))

#plot distribution of categorical data with count plot
sns.countplot(x=education_quality)
```
