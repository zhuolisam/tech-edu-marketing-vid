```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset with pandas.
df = pd.read_csv('../Data/StudentsPerformance.csv')

#Use head() function to see the first 5 rows of the dataset.
df.head()
```

```python
plt.figure(figsize=(12,6))

#Fit in the columns of data by specifying the column names, and pass in the dataframe
#Visualize the distribution of math score
sns.distplot(x='math score', kde=True, data=df)
```

```python
plt.figure(figsize=(12,6))

#Plot a barplot with the mean of the math score 
sns.barplot(x='parental level of education',y='math score',data=df)
```

```python
plt.figure(figsize=(12,6))

# We can also see how male and female students differs by using hue parameter
sns.barplot(x='parental level of education',y='math score',hue='gender', data=df)
```
