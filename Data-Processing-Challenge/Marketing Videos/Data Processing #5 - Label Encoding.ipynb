```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Why do we need to encode categorical variables?
- Machine learning algorithms cannot work with categorical data directly. 

- The categories must be converted into numbers. 

- This is required for both input and output variables that are categorical.

```python
#import dataset
df = pd.read_csv('../Data/titanic.csv')
df.head()
```

```python
df_categorical = df[['Sex']]
df_categorical.head()
```

```python
#Use One Hot Encoder on input categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

onehotencoder = OneHotEncoder()
df_encoded = pd.DataFrame(onehotencoder.fit_transform(df_categorical[['Sex']]).toarray(), columns =['Male','Female'])

df_encoded = df_categorical.join(df_encoded)
df_encoded
```

```python
#Use Label Encoder on output categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()

df_survive = df[['Survived']]

df_survived_encoded = pd.DataFrame(labelencoder.fit_transform(df_survive['Survived']), columns =['Output'])
df_survive.join(df_survived_encoded)
```
